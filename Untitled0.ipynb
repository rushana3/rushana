Ввод и вывод информации в консоль. Объекты и числа. Переменная - простейшая именованная структура данных, в которой может быть сохранён промежуточный или конечный результат работы программы. Строка - это то, что пишется в кавычках ' ' или " ". greeting это имя переменной. Переменная это хванилище данных. По правилам имя может состоять из латинских букв, цифр и подчеркиваний, может начинаться с цифры. Имя не должно содержать зарезервированные слова "False, break, true, else". Принцип работы переменных: Создается строка (str), создается переменная в переменной greeting создается адрес на объект. В пайтоне всё состоит из объектов. = это знак присваивания, присваиваем значение справа к переменной слева. Значение всегдла относится к какому-либо типу данных Это str. Тип можно узнать через функцию type. 2 строка.

```python
greeting = "Hello, world"
print(type(greeting))
```

Адрес это номер ячейки памяти, в которой хранится значение переменной. Адрес можно посмотреть с помощью формулы ID. Переменная дает возможность удобно работать с объектами и пользоваться ими

```python
print(id(greeting))
```

print это функция которая может выводить информацию. Мы можем вывести строку Hello, world. Также можно добавить какие-либо параметры в "" Это вывод. Greeting - первый параметр my name - второй. Вывели строку гритинг

```python
print(greeting, 'my name is Rushana')
```

Удобно работать с F-строками. f строка это просто некоторый комфортный вывод информации. С помощью функции help можно расмотреть некоторые аргументы, которые есть в этой функции. Если хотим, чтобы элементы разделялись через равно

Именованный аргумент END - это то, чем заканчивается вывод. По умолчанию End равен back slash n ( \n ) - переход на новую строку

```python
name = 'Rusha'
print('My name', end='\n')
print(name)
```

```python
name = 'Rusha'
print('My name', end=':')
print(name)
```

f формула. В фигурных скобках переменную указывать

```python
name = 'Rushana'
print(f'My name is {name}')
```

```python
age = 18
print(f' I am {age +1}')
```
